```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# Coordinates (update these with your actual values from previous steps)
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# --- Example launch site coordinates and info (update these for your real data) ---
launch_sites_info = [
    {'name': 'CCAFS SLC-40', 'lat': 28.563197, 'lon': -80.576820, 'count': 26, 'color': '#d35400'},
    {'name': 'CCAFS LC-41',  'lat': 28.573255, 'lon': -80.646895, 'count': 7,  'color': '#27ae60'}
]

# Your selected coastline coordinates (use MousePosition to get the exact values)
coastline_lat = 28.565    # Example: set with MousePosition result
coastline_lon = -80.568   # Example

# --- Distance calculation function (given as provided) ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# --- Create map centered on first launch site ---
center_lat = launch_sites_info[0]['lat']
center_lon = launch_sites_info[0]['lon']
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=16)

# --- Add colored circles and labels for each launch site ---
for site in launch_sites_info:
    # Circle
    folium.Circle(
        [site['lat'], site['lon']],
        radius=110,
        color=site['color'],
        fill=True,
        fill_color=site['color'],
        fill_opacity=0.25
    ).add_to(site_map)
    # Label (DivIcon) with count and site name
    label_html = f"""<div style="font-size: 14px; color: #ffffff; background-color: {site['color']}; border-radius: 20px; width: 40px; text-align: center;">
                  <b>{site['count']}</b><br><span style="font-size: 11px;">{site['name']}</span></div>"""
    folium.Marker(
        [site['lat'], site['lon']],
        icon=DivIcon(
            icon_size=(40,40),
            icon_anchor=(20,20),
            html=label_html
        )
    ).add_to(site_map)

# --- Draw PolyLine from closest launch site to coastline ---
# (Choose site by proximity OR use first site; update as needed)
launch_site_lat = launch_sites_info[0]['lat']
launch_site_lon = launch_sites_info[0]['lon']
line_coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
folium.PolyLine(
    line_coordinates,
    color='blue',
    weight=2,
    opacity=1
).add_to(site_map)

# --- Add coastline marker with orange distance label ---
distance_km = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(60,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 14px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_km)
    )
).add_to(site_map)

# --- Show the map ---
site_map

```
